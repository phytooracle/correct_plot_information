```python
import os
import sys
import geopandas as gpd
```

```python
new_gdf = gpd.read_file('./sorghum/season14_multi_latlon_geno_correction.geojson')
old_gdf = gpd.read_file('./sorghum/season14_multi_latlon_geno_updated.geojson')
```

```python
old_gdf = old_gdf[['ID', 'geometry']].rename(columns={
    'ID': 'error_ID',
})
```

```python
gdf = new_gdf.merge(old_gdf, on='geometry')
```

```python
gdf.loc[gdf['ID'] != gdf['error_ID'], 'mismatch'] = 'error'
gdf.loc[gdf['ID'] == gdf['error_ID'], 'mismatch'] = 'no error'
```

```python
fb = pd.read_excel('https://data.cyverse.org/dav-anon/iplant/projects/phytooracle/season_14_sorghum_yr_2022/North_gantry_fieldbook_2022_replants.xlsx',
                   sheet_name='Fieldbook_replant')

fb = fb[['year', 'range', 'row', 'plot', 'type', 'rep', 'treatment', 'species', 'accession', 'Entry_ID']]
fb['plot'] = fb['plot'].astype(str)
```

```python
gdf['plot'] = gdf["ID"]
```

```python
gdf = fb.merge(gdf, on='plot')
```

```python
gdf = gpd.GeoDataFrame(gdf)
```

```python
gdf.to_file('./sorghum/season14_multi_latlon_geno_correction_labeled.geojson', driver="GeoJSON")  
```
